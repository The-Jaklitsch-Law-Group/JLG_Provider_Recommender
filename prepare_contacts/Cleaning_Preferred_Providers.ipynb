# Cleaning Contact Data - Preferred Providers

```python
import pandas as pd
```

```python
df = pd.read_excel('../data/raw/Referral_App_Preferred_Providers.xlsx')
df = df.drop_duplicates(ignore_index=True)
df
```

```python
df["Contact's Details: Latitude"].notna().mean().round(2)
```

```python
df[df["Contact's Details: Latitude"].isna()].sort_values(by= ' Full Name', ignore_index=True)#.to_excel('../data/processed/Preferred_Providers_Missing_LatLong.xlsx', index=False)
```

```python
df = df.dropna(subset=["Contact's Details: Latitude", "Contact's Details: Longitude"],
               how='any',
               axis=0,
               inplace=False,
               ignore_index=True
               )
df
```

```python
df.to_parquet('../data/processed/cleaned_preferred_providers.parquet', index=False)
```
