```python
df = pd.read_excel('./data/List of Contacts - Provider Referrals.xlsx',
                    parse_dates = ['Create Date',
                                    'Date of Intake',
                                    'Sign Up Date'],
                    dtype = {'Dr/Facility Referred To Address 1 Zip':str})
df
```

```python
## Assume that the "Sign Up Date" is most accurate, then "Date of Intake", else "Create Date"

df['Referred Date'] = df['Sign Up Date'].fillna(value = df['Date of Intake'])
df['Referred Date'] = df['Referred Date'].fillna(value = df['Create Date'])
df['Referred Date'].isna().sum()
```
